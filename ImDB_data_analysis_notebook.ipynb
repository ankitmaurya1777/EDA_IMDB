# Q3. How do we clean the `duration` column? (e.g., `'175 min'` → `175`)


```python
df["duration"]= df["duration"].str.replace('min', "")
```

```python
df["duration"] = df["duration"].str.replace(",","")
```

```python
df["duration"] = df["duration"].astype(int)
```

```python
df.head()
```

# Q4. How do we clean the `gross` column and `votes` column?


```python
df["gross_income"] = df["gross_income"].str.replace("," , "")


```

```python
df["gross_income"] = df["gross_income"].str.replace("$" , "")
df["gross_income"] = df["gross_income"].str.replace("." , "")
df["gross_income"] = df["gross_income"].str.replace("M" , "")
```

```python
df["gross_income"] = df["gross_income"].astype(float)
```

```python
df.head()
```

```python
df["votes"] = df["votes"].str.replace(",","")
```

```python
df["votes"] = df["votes"].astype(int)
```

```python
df.head()
```


# Q22. Which Director + Actor duo is the most frequent?

```python
duo_df = df[['directors_name', 'stars_name']].copy()


duo_df['directors_name'] = duo_df['directors_name'].str.split(',')
duo_df = duo_df.explode('directors_name')


duo_df['stars_name'] = duo_df['stars_name'].str.split(',')
duo_df = duo_df.explode('stars_name')


duo_df['directors_name'] = duo_df['directors_name'].str.strip()
duo_df['stars_name'] = duo_df['stars_name'].str.strip()


top_10_duos = duo_df.groupby(['directors_name', 'stars_name']).size().reset_index(name = "count")
top_10_duos = top_10_duos.sort_values(by  = "count", ascending=False).head(10)

```

```python
top_10_duos
```

```python
plt.figure(figsize=(12,6))

top_10 = top_10_duos["stars_name"] + "+" + top_10_duos["directors_name"]


plt.barh(top_10, top_10_duos["count"], color = "red")
plt.xlabel("most frequent times")
plt.ylabel("name of duos")

plt.title("most frequent actor + directoer duos")
plt.grid(alpha=0.4)
plt.show()
```

# Q28. The Ultimate Business Decision: What movie should we fund?

```python
genre_rating = high_rating.groupby("genre")["rating"].mean()
genre_gross = high_rating.groupby("genre")["gross_income"].mean()

strategy = pd.concat([genre_rating, genre_gross], axis=1)
strategy.columns = ["avg_rating","avg_gross"]

strategy.sort_values(["avg_rating","avg_gross"], ascending=False).head()
```
